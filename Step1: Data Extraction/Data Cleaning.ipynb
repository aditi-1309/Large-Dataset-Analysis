```python
# Reading the CSV file with the dataset into stack_exchange
stack_exchange = pd.read_csv("D:\DCU\Academics\CA675 Cloud Tech\Assignment\Final_DataSet.csv")
```

```python
# Dropping columns ParentId and DeletionDate as they only have null values
to_drop = ['ParentId', 'DeletionDate']
stack_exchange.drop(columns=to_drop,inplace=True, axis=1)
```

```python
# Checking if the columns are deleted properly
stack_exchange.info()
```

```python
# Importing regular experssion package
import re

# Function to remove HTML tags from the data
def remove_tag(data):
    clean_data = re.sub(pattern=re.compile('<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});|\.*?'),repl='', string=str(data))
    return clean_data
```

```python
# Creating a list by removing HTML tags from Title column
no_tag_title=[]
for i in range(len(stack_exchange.Title)):
    no_tag_title.append(remove_tag(stack_exchange.Title[i]))
```

```python
# Creating a list by removing HTML tags from Body column
no_tag_body=[]
for i in range(len(stack_exchange.Body)):
    no_tag_body.append(remove_tag(stack_exchange.Body[i]))
```

```python
print(no_tag_body[20])
print(stack_exchange.Body[20])
```

```python
# Dropping the columns Body and Title
stack_exchange.drop(['Body','Title'], axis = 1, inplace = True)
```

```python
# Creating the columns Body and Title without the HTML tags
stack_exchange['Body'] = no_tag_body
stack_exchange['Title'] = no_tag_title

# Removing escape characters from Body and Title columns
stack_exchange['Body'] = stack_exchange['Body'].str.replace(r'\n|\t|\r','',regex=True)
stack_exchange['Title'] = stack_exchange['Title'].str.replace(r'\n|\t|\r','',regex=True)
```
